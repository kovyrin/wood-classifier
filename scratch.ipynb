```python
from pathlib import Path

# Cleaning up the hobbithouseinc dataset
data_dir = "data/hobbithouseinc-clean/grain"
data_dir = Path(data_dir)

# For each wood species in the data folder
for wood_species in data_dir.iterdir():
    print("Checking for duplicates in", wood_species.name)
    images = list(wood_species.iterdir())
    image_names = [image.name for image in images]

    # Check if we have a higher resolution image (low res are named s25.jpg, while high res are named s50.jpg or s100.jpg)
    for image in image_names:
        if "s25" in image:
            potential_high_res_image = image.replace("s25", "s50")
            if potential_high_res_image in image_names:
                print("Found a better version for {}, using {} instead".format(image, potential_high_res_image))
                # Delete the low res image
                (wood_species / image).unlink()
        if "s50" in image:
            potential_high_res_image = image.replace("s50", "s100")
            if potential_high_res_image in image_names:
                print("Found a better version for {}, using {} instead".format(image, potential_high_res_image))
                # Delete the low res image
                (wood_species / image).unlink()

```
